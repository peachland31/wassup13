### 인덱싱과 슬라이싱, 데이터 선택
파이썬의 인덱스는 기본적으로 0부터 시작한다.  
예를 들어, a = ['x', 'y', 'z']의 첫번째 인덱스인 x를 호출하기 위해 0번 인덱스를 조회한다.  
a[0] = 'x', a[1] = 'y', a[2] = 'z' (0번, 1번, 2번 Index 호출)  

* colab(1) 파이썬 기초 - '인덱스 구조' 참고

````python
# 문자열, 리스트 등의 기본 인덱싱

a = ['x', 'y', 'z']
print(a[0])
print(a[1])
print(a[2])
````

````python
# 슬라이싱의 기본 개념 a[시작:끝 -1] (디폴트값은 전체)

a = ['x', 'y', 'z']
print(a[0:2]) # 0번부터 1번까지 인덱싱
print(a[:3]) # 2번까지 인덱싱
print(a[1:]) #  1번부터 인덱싱
print(a[:]) # 전체 인덱싱
````

````python
# 기본 인덱싱 0부터 시작, 슬라이싱은 끝 -1 적용, 데이터 프레임의 기본 축은 axis = 행 0, 열 1

# 칼럼을 선택하는 가장 기본적인 방법
df['PassengerId'] # df.PassengerId 방식으로도 선택할 수 있지만, 코딩 편의상 [] 형식으로 통일.
````

#### 불리언(boolean) 인덱싱  
불리언 연산자에 따라 True, False 값으로 분류하여 조회하는 방식  
데이터 프레임 형식으로는 df[df['칼럼'] 불리언 연산자]를 사용한다.

많이 쓰이는 불리언 연산자:
<, >, ==, !=, |, &

````python
# Survived가 1인 데이터 조회
df[df['Survived'] == 1].head() # df[df['Survived'] != 1].head()를 입력하면 반대의 결과가 출력 된다.(== <-> !=)
````

````python
# Fare가 100이상인 데이터 조회
df[df['Fare'] >= 100].head()
````

````python
# Fare가 100이상이거나, Survived가 1인 데이터 (OR 연산자)
df[(df['Fare'] >= 100) | (df['Survived'] == 1)].head()
````

````python
# Fare가 100이상이면서, Survived도 1인 데이터 (AND 연산자)
df[(df['Fare'] >= 100) & (df['Survived'] == 1)].head()
````

## 결측치 중앙값으로 대체하기


````python
# df['Age']의 중앙값 조회
df['Age'].median()
````

````python
# Age칼럼의 결측치 개수 확인
df[['Age']].isna().sum()
````

````python
# sort_values로 재확인
# 총 row 891에서 결측치 177을 제외하면 총 714개의 행이된다. 즉, 714개의 행에다한 중앙값을 계산하면 된다.
(df['Age'].dropna().sort_values(ascending=False).reset_index(drop=True)[356] + df['Age'].dropna().sort_values(ascending=False).reset_index(drop=True)[357]) / 2
````

````python
# fillna로 결측치를 중앙값으로 대체
df_2 = df.copy()
df_2['Age'] = df_2['Age'].fillna(df_2['Age'].median())
print(df_2[['Age']].isna().sum())
df_2[['Age']].value_counts(ascending=False, dropna=False)
````

중앙값, 평균값, 최빈값 등 하나의 값으로 결측치를 처리한다면, 데이터 분포가 한쪽에 쏠릴 수 있기 때문에  
데이터에 따라, 다양한 코드와 라이브러리를 활용한다.

````python
# 심화 패턴: Age에 가장 영향을 많이 미친 Sex, Pclass 그룹당 median를 구하여 해당 값을 결측치로 대체

df.groupby(['Sex', 'Pclass'])['Age'].median()
````

````python
# 정답이라고 할 수 있는 방법은 없다. 다양한 기법과 라이브러리들이 있으니 데이터에따라 적절한 접근 방식을 설계 해보는 것이 필요하다.
df_3 = df.copy()
df_3['Age'] = df_3['Age'].fillna(df_3.groupby(['Sex', 'Pclass'])['Age'].transform('median'))
df_3['Age'].value_counts(ascending=False, dropna=False)
````

````python
df_3[['Age']].isna().sum()
````

## 새로운 Feature 생성  
결측치가 있던 행들은 분석시 혼선을 줄 수 있으므로, 새로운 결측치 여부 Feature로 만들어 구분할 수 있다.  
또한, 기존의 Feature를 기반으로 다양한 Feature Engineering을 수행 할 수 있다.

````python
# 결측치가 가장 많은 객실정보 Cabin은 'unknown'으로 대체하고, Embarked는 결측치가 2개이므로 간단하게 최빈값으로 대체한다.

df_3['Cabin'] = df_3['Cabin'].fillna('unknown')
df_3['Embarked'] = df_3['Embarked'].fillna(df_3['Embarked'].mode()[0])
````

````python
# Age_null 칼럼을 따로 만들어 두어, 결측치 여부를 판단할 수 있게 한다. (0, 1) 이진 분류

df_3['Age_null'] = df['Age'].isna().astype(int)
df_3
````

### 새로운 라벨을 만들어 나이 카테고리 Feature 생성하기  
Age와 같은 중요하면서도 세그먼트가 나뉘는 데이터들은 분석에 따라 연속형 수치가 아닌 범주형으로 보는 것이 더 좋을 수 있다.

````python
# 나이를 구간화, 카테고리화한 변수 지정하기

age_bins = [0, 12, 19, 34, 58, 100]
age_cat = ['Child', 'Teen', 'Young', 'Adult', 'Senior']

print(age_bins, age_cat)
````

````python
# pd.cut으로 라벨링 된 데이터 만들기

AgeGroup = pd.cut(df_3['Age'], bins=age_bins, labels=age_cat)
AgeGroup.head()
````

````python
# 새로운 Feature인 AgeGroup 추가
df_3['AgeGroup'] = AgeGroup
df_3
````

# 데이터 정제 및 분석 점검  
몇 가지 핵심적인 정제를 마친 df_3로 어떤 분석을 할 수 있는지 연습 해보기

````python
# 이름에 Mrs.를 포함하면서 생존한 사람 조회

df_3_Mrs = df_3[(df_3['Name'].str.contains('Mrs') & df_3['Survived'] == 1)].reset_index(drop=True)
df_3_Mrs.head()
````

````python
# 생존한 기혼 여성은 대부분 Adult, Young 그룹에 속해있다.
df_3_Mrs[['AgeGroup']].value_counts()
````

````python
# 전체 연령 그룹을 확인했을때, 청년, 성인그룹이 가장 많이 분포되어 있다.
df_3['AgeGroup'].value_counts()
````

````python
# 객실 등급은 3등석이 가장 많았고, 2등석이 제일 적었다.
df_3['Pclass'].value_counts()
````

````python
# 결측치가 많았던 Age, 영향 있었을까? = Age 값이 결측치였던 삶들의 생존율
df_3.groupby("Age_null")["Survived"].mean()
````

````python
# 생존한 객실 등급 그룹은 높을수록 생존율이 좋았다.
df_3.groupby("Pclass")["Survived"].mean().round(2)
````

````python
# 생존한 성별 그룹을 볼 때, 성별은 여성 우선, 남성은 아이 우선으로 살리려고 한 것을 유추할 수 있다.
df_3.groupby(["Sex","AgeGroup"])["Survived"].mean().round(2)
````

````python
# 객실 등급이 높을수록 전체적인 생존율이 높고, 나이가 적은 그룹이 우선적으로 생존된 걸 확인할 수 있다.
df_3.groupby(["Pclass", "AgeGroup"])["Survived"].mean().round(2)
````

````python
# 단순 비율뿐만 아니라, 승객 인원수에 따른 생존율과 평균적인 객실 요금 확인이 가능하다.
df_3.groupby(["Sex", "Pclass"]).agg(cnt_Id=("PassengerId", "count"),avg_survived=("Survived", "mean"),avg_fare=("Fare", "mean")).round(2)
````
